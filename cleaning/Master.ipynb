```python
import numpy as np
import pandas as pd
```

```python
dfs=[]
cols=['Age', 'Sex','Marital status at diagnosis', 'Year of diagnosis',           #list of columns/variables used for analysis
       'Race and origin recode (NHW, NHB, NHAIAN, NHAPI, Hispanic)','Primary Site',
       'Histologic Type ICD-O-3','Stage','Treatment','Survival months',
       'Vital status recode (study cutoff used)','COD to site recode',
       'Rural-Urban Continuum Code','Cancer_type',
        'Income']
csvs=[
    'Breast_Clean.csv','Liver_Clean.csv','Lung_Clean.csv','Prostate_Clean.csv','Stomach_Clean.csv' #list of csv's 
]
dfs=[pd.read_csv(i,usecols=cols) for i in csvs] #filtering columns from each clean data csv
clean=pd.concat(dfs,ignore_index=True)
```

```python
clean = clean.drop(columns=['Unnamed: 0'], errors='ignore') #sabing all into one csv
clean.to_csv('FinalData.csv')
```

```python
df = pd.read_csv('FinalData.csv')
```

```python
cancer_map = {
    'Breast':            'BreastClean',
    'Liver':             'LiverClean',
    'Lung and Bronchus': 'LungClean',
    'Prostate':          'ProstateClean',
    'Stomach':           'StomachClean',
}
sheet_names = ['I', 'II', 'III', 'IV']
for c,f in cancer_map.items(): #creating stage wise sheets in cancer wise excel workbooks
    df1=df[df['Cancer_type']==c]
    print(1)
    file_name=f'{f}.xlsx'
    with pd.ExcelWriter(file_name, engine='openpyxl') as writer:
        for s in sheet_names:
            df2=df1[df1['Stage']==s]
            print(2)
            df2.to_excel(writer,sheet_name=f'stage_{s}',index=False)
            print(3)
    print(f'{f}.xlsl done!')
```
